```python
import numpy as np
import pandas as pd
```

```python
#Cargamos los datasets, y creamos la columna "id" que consta de la primer letra del nombre de la plataforma + el show_id
#Asignamos a cada plataforma una columna "platform" con el nombre de la plataforma

Disney = pd.read_csv('MLOpsReviews/disney_plus_titles.csv')
Disney['id'] = 'd' + Disney['show_id'].astype(str)
Disney['platform'] = 'disney'

Amazon = pd.read_csv('MLOpsReviews/amazon_prime_titles.csv')
Amazon['id'] = 'a' + Amazon['show_id'].astype(str)
Amazon['platform'] = 'amazon'

Netflix = pd.read_csv('MLOpsReviews/netflix_titles.csv')
Netflix['id'] = 'n' + Netflix['show_id'].astype(str)
Netflix['platform'] = 'netflix'

Hulu = pd.read_csv('MLOpsReviews/hulu_titles.csv')
Hulu['id'] = 'h' + Hulu['show_id'].astype(str)
Hulu['platform'] = 'hulu'
```

```python
#Creamos un solo dataset para optimizqar el trabajo
data = pd.concat([Disney, Amazon, Netflix, Hulu])
data = data[data.columns[-1:].tolist() + data.columns[:-1].tolist()]
data['rating'] = data['rating'].fillna('G')
```

```python
#Convertimos la fecha de la columna "data_added" a un formato AAAA-mm-dd
data['date_added'] = data['date_added'].str.strip()
data['date_added'] = pd.to_datetime(data['date_added'], format='%B %d, %Y')
data['date_added'] = data['date_added'].dt.strftime('%Y-%m-%d')
```

```python
#Dividimos la columna "duration" en "duration_type" y "duration_int" que contendran el tipo de duracion y su cantidad respectivamente
data[['duration_int', 'duration_type']] = data['duration'].str.extract('(\d+) (\w+)')
data['duration_type'] = data['duration_type'].replace('Seasons', 'Season')
data = data.drop('duration', axis=1)
```

```python
#Pasamos todo texto a minuscula
data[['type', 'title', 'director', 'cast', 'country', 'listed_in', 'description', 'duration_type']] = data[['type', 'title', 'director', 'cast', 'country', 'listed_in', 'description', 'duration_type']].applymap(lambda x: x.lower() if type(x) == str else x)
```

```python
#renombramos "id" como "movieId" para estar en concordancia con los demas datasets
data.rename(columns={'id': 'movieId'}, inplace=True)
data
```

```python
#Cargamos los datasets de calificaciones
rating_1 = pd.read_csv('MLOpsReviews/ratings/1.csv')
rating_2 = pd.read_csv('MLOpsReviews/ratings/2.csv')
rating_3 = pd.read_csv('MLOpsReviews/ratings/3.csv')
rating_4 = pd.read_csv('MLOpsReviews/ratings/4.csv')
rating_5 = pd.read_csv('MLOpsReviews/ratings/5.csv')
rating_6 = pd.read_csv('MLOpsReviews/ratings/6.csv')
rating_7 = pd.read_csv('MLOpsReviews/ratings/7.csv')
rating_8 = pd.read_csv('MLOpsReviews/ratings/8.csv')
```

```python
#Concatenamos los dataset para trabajar con uno solo
rating = pd.concat([rating_1, rating_2, rating_3, rating_4, rating_5, rating_6, rating_7, rating_8])
rating
```

```python
#Calculamos el promedio de callificacion de cada pelicula
promedios = rating.groupby('movieId')['rating'].mean()
promedios
```

```python
#Agregamos el promedio de la calificacion al primer dataset
data = pd.merge(data, promedios, on='movieId')
data.rename(columns={'rating_y': 'score'}, inplace=True)
data.rename(columns={'rating_x': 'rating'}, inplace=True)
data['score'] = data['score'].astype(float)
data
```

```python
#Buscamos titulos duplicados
duplicated = data[data.duplicated(['title'], keep='first')]

# Eliminar todas las filas duplicadas excepto la primera
data.drop_duplicates(['title'], keep='first', inplace=True)
print("Títulos de filas eliminadas:\n", duplicated['title'])
```

```python
#Buscamos los balores unicos de la calificacion etaria
data["rating"].unique()
```

```python
#Establecemos los valores que aceptamos
valores_permitidos = ['TV-G', 'PG', 'TV-PG', 'G', 'PG-13', 'TV-14', 'TV-Y7', 'TV-Y',
                      'TV-Y7-FV', '13+', 'ALL', '18+', 'R', 'NR', '16+', '7+', 'TV-NR',
                      'TV-MA', 'NC-17', 'UNRATED', '16', 'AGES_16_', 'AGES_18_',
                      'ALL_AGES', 'UR']

# Reemplazar los valores que no estén permitidos con "unrated"
data['rating'] = data['rating'].apply(lambda x: 'unrated' if x not in valores_permitidos else x)
data["rating"].unique()
```

```python
#Estandarizamos cirtos valores similares
data['rating'] = data['rating'].replace('16', '16+')
data['rating'] = data['rating'].replace('AGES_16_', '16+')
data['rating'] = data['rating'].replace('AGES_18_', '18+')
data['rating'] = data['rating'].replace('unrated', 'UNRATED')
data["rating"].unique()
```

```python
data.to_csv('datos_limpios.csv', index=False)
```
